```python
import pandas as pd
import numpy as np
from pykrx import stock
```

## 연간 컨센선스

```python
data_path = '/content/drive/MyDrive/프로젝트/어닝/data/processed data/'
```

```python
con_year = pd.read_csv(data_path + 'con_year.csv', header=[0,1], index_col=[0,1])
```

```python
year = np.arange(2018, 2022)
col = ['매출액', '영업이익', '세전계속사업이익', '순이익', '지배주주귀순순이익']

con_year.columns = pd.MultiIndex.from_product([year, col])
con_year = con_year.reset_index(1).rename(columns={'level_1':'name'})
```

```python
kospi200 = stock.get_index_portfolio_deposit_file("1028") #코스피200종목 리스트
kospi200
name_list = []
for ticker in kospi200:#코스피200종목명 리스트
        name = stock.get_market_ticker_name(ticker)
        name_list.append(name)
```

```python
con_year = con_year[con_year.name.apply(lambda x:x in name_list)]
con_year.to_csv('/content/drive/MyDrive/프로젝트/어닝/data/final data/con_year.csv')
con_year.head() #단위 억원
```

```python
set(name_list).difference(con_year.name) # 컨센선스 없는 기업 
```

## 연간 매출

```python
sales_year = pd.read_csv(data_path + 'sales_year.csv', header=[0,1], index_col=[0,1])
```

```python
year = np.arange(2018, 2020)
col = ['매출액', '매출총이익', '판관비', '조정영업이익', '당기순이익', '지배주주귀순순이익']

sales_year.columns = pd.MultiIndex.from_product([year, col])
sales_year = sales_year.reset_index(1).rename(columns={'level_1':'name'})
```

```python
sales_year = sales_year[sales_year.name.apply(lambda x:x in name_list)]
for col in sales_year.columns[1:]:
  sales_year[col] = sales_year[col].apply(lambda x: float(x.replace(',',''))/100000 if type(x)==str else x)
sales_year.to_csv('/content/drive/MyDrive/프로젝트/어닝/data/final data/sales_year.csv')
sales_year.head() #단위 억원
```
